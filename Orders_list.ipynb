```python
from pandas.core.groupby.groupby import DataFrame
import pandas as pd
import json
import itertools as it
from IPython.display import Image, HTML
import requests
import time

def flatten_json (dictionary):

    def unpack(parent_key,parent_value):
        if isinstance (parent_value,dict):
            for key, value in parent_value.items():
                temp = parent_key + "-" + key
                #print("parent_key: ", parent_key," key: ", key)
                yield temp, value
        elif isinstance(parent_value, list):
            i = 0
            for value in parent_value:
                temp2 = parent_key + "_" + str(i)
                i += 1
                yield temp2, value
        else:
            yield parent_key, parent_value
    while True:
        dictionary = dict(it.chain.from_iterable(it.starmap(unpack,dictionary.items())))

        if not any(isinstance(value,dict) for value in dictionary.values()) and not any(isinstance(value,list) for value in dictionary.values()):
            break

    return dictionary



f = open('/content/liv100')
json_file = json.load(f)
#f2 = open('/content/221126')
#json_file2 = json.load(f2)

results = json_file["results"]
#results2 = json_file2["results"]

#results.extend(results2)

def unique(sequence):
    seen = set()
    return [x for x in sequence if not (json.dumps(x) in seen or seen.add(json.dumps(x)))]
newOBJ = {}
newArr = []

def getImageLink(listingID, listingImgID):
    response = requests.get("https://openapi.etsy.com/v2/listings/" + listingID +"/images/" + listingImgID + "?api_key=7hvm9ahb1pepq3lybzs5f3l6")
    try:
      if (response.status_code == 200):
          #print("The request was a success!")
          # Code here will only run if the request is successful
          url = response.json()["results"][0]['url_170x135']
          return f'<img src="{url}"/>'
          time.sleep(0.11)
    except:
      
          # Code here will react to failed requests
          print("Result not found!")
          #print(response.json())
          return "image could not be found"
    
counter = 0
for reciept in unique(results):

  new = {
      "transactions": [{"title": item["title"], "quantity": item["quantity"], "sku": item["sku"], "variations": item["variations"], "photo": getImageLink(str(item["listing_id"]), str(item["listing_image_id"]))} for item in reciept["transactions"]],
      "is gift": reciept["is_gift"],
      "Country Code": reciept["country_iso"],
      
  }
  newArr.append(new)
  newOBJ[reciept["name"]] = new
  counter += 1
  print(counter)
newArr
#df = pd.json_normalize(newArr, record_path=['transactions'], meta=['name', 'is gift'])
#df.to_excel("cars_data_9.xlsx")

dictionary = flatten_json(newOBJ)
all_values = list(dictionary.values())

index_list = []
for key in dictionary:
    x = tuple(key.split("-"))
    index_list.append(x)

index = pd.MultiIndex.from_tuples(index_list)

df = pd.Series(all_values, index=index).to_frame()
def path_to_image_html(path):
    return '<img src="'+ path + '"/>'
pd.set_option('display.max_colwidth', None)

def color(val):
    if val < 1:
        color = 'green'
    elif val > 1:
        color = 'red'
    return 'background-color: %s' % color
df.style.applymap(color, subset=['0'])
df
HTML(df.to_html(escape=False))
df.to_html('261122_bıçak.html', escape=False)
#df.to_excel("191122_siparişler.xlsx")
```
